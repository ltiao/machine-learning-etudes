```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
```

```python
golden_ratio = 0.5 * (1 + np.sqrt(5))
golden_size = lambda width: (width, width / golden_ratio)
```

```python
width_pt = 3 * 234.8775 # Get this from LaTeX using \showthe\columnwidth
inches_per_pt = 1.0/72.27 # Convert pt to inches
width = width_pt * inches_per_pt # width in inches

rc = {
    "figure.figsize": golden_size(width),
    "font.serif": ['Times New Roman'],
    "text.usetex": True,
}

sns.set(context="notebook", style="ticks", palette="colorblind", font="serif", rc=rc)
```

```python
data = pd.read_csv("../datasets/co2_mm_mlo.txt", 
                   names=["year", "month", "date", "average", "interpolated", "trend", "num_days"],
                   comment="#", header=None, sep="\s+")
data = data[data.average > 0]
```

```python
g = sns.relplot(x='date', y='average', kind="line",
                data=data, height=5, aspect=golden_ratio, alpha=0.8)
g.set_ylabels(r"average $\mathrm{CO}_2$ (ppm)")
```
